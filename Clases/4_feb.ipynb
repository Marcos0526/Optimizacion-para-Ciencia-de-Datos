```python
def biseccion(f, a, b, e):
    while b-a>=e:
        c=(a+b)/2
        if f(c)==0:
            return c
        else:
            if f(a)*f(c)>0:
                a=c
            else:
                b=c
    return c
```

```python
from math import*
def f(x):
    return x*exp(x)-pi
c=biseccion(f,0,2,0.000001)
c
```

```python
import math
def newton_method(f, df, x0, tol=1e-6, max_iter=100):
    print(f"{'Iter':<5} | {'x_n':<10} | {'f(x_n)':<10}")
    print("-" * 35)
    for i in range(max_iter):
        fx = f(x0)
        print(f"{i:<5} | {x0:<10.6f} | {fx:<10.6e}")
        if abs(fx) < tol:
            return x0
        x0 = x0 - fx / df(x0)
    return x0
```

```python
def f15(x):
    return 5*x - math.exp(x) - 1

def df15(x):
    return 5 - math.exp(x)
```

```python
print("Buscando primera raz (cerca de 0):")
r1 = newton_method(f15, df15, 0)
print(f"\nRaz 1 encontrada: {r1:.6f}\n")

print("Buscando segunda raz (cerca de 3):")
r2 = newton_method(f15, df15, 3)
print(f"\nRaz 2 encontrada: {r2:.6f}")
```

```python
A = math.e / 3

def f16(t):
    return A * t * math.exp(-t/3) - 0.25

def df16(t):
    return A * math.exp(-t/3) * (1 - t/3)

print("Buscando tiempo t donde C(t) = 0.25 (despus del mximo, x0=10):")
t_final = newton_method(f16, df16, 10, tol=1e-4)
print(f"\nEl tiempo t es: {t_final:.6f}")
```
